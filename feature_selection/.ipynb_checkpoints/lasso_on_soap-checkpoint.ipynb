```python
import numpy as np
import os
import pickle

my_dir = '/content/soap_amr_data'
files = np.sort(os.listdir(my_dir))
gb_list = []
for file in files:
    with open(my_dir+'/'+file,'rb') as f:
        data = pickle.load(f)
        gb_list.append(data)

energy = np.load('/content/drive/MyDrive/small_energy.npy')
```

```python
print(f'(Number of grain boundaries,length of SOAP representation): {np.shape(gb_list)}')
print(f'length of energies: {len(energy)}')
```

```python
import numpy as np
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score

# Data
X = np.array(gb_list)
y = energy

# Normalize the data using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Range of lambda values to try
lambdas = np.logspace(-3.5, 2.5, 60)

# Tolerance for considering coefficients as zero
tolerance = 1e-5

# List to keep track of the number of non-zero coefficients
non_zero_counts = []
costs = []
r2_scores = []
image_of_nonzero_params = np.zeros((np.shape(X)[1],len(lambdas)))
# Loop over the lambda values
for i,l in enumerate(lambdas):
    # Create and fit the Lasso model
    model = Lasso(alpha=l)
    model.fit(X_scaled, y)

    # Count the number of non-zero coefficients (considering the tolerance)
    image_of_nonzero_params[:,i] = np.abs(model.coef_) > tolerance
    non_zero_count = np.sum(np.abs(model.coef_) > tolerance)
    non_zero_counts.append(non_zero_count)
    y_pred = model.predict(X_scaled)
    costs.append(np.linalg.norm(y_pred-y))
    r2_scores.append(r2_score(y, y_pred))


# Output the lambda values and corresponding non-zero counts
for l, count in zip(lambdas, non_zero_counts):
    print(f"Lambda: {l:.2e}, Non-Zero Coefficients: {count}")

```

```python
# Train the Lasso model with the optimal lambda value (10.0)
optimal_lambda = 17.0
optimal_model = Lasso(alpha=optimal_lambda)
optimal_model.fit(X_scaled, y)

# Extracting the coefficients
coefficients = optimal_model.coef_

# Identifying the non-zero coefficients and mapping them to the features
non_zero_coefficients = np.where(coefficients != 0)[0]
important_features = non_zero_coefficients

important_features, coefficients[non_zero_coefficients]
```

```python
# Decode the important feature indeces
import numpy as np

Nmax = 12 # values taken from the original SOAP code that Braxton used
Lmax = 12

# Logic for decoder

def decl(F): # decode the l component
  N = Nmax*(Nmax+1)/2
  Fp = F%N
  l = (F-Fp)/N
  return [int(l),Fp]

def decn(F): # decoder for n
  ns = np.arange(1,Nmax+1) # all candidate n values
  sms = np.zeros_like(ns) # empty array for summing
  for n in ns:
    sms[n-1] = sum(np.arange(n,Nmax+1)) - F # perform the summation

  K = np.where(sms==min(sms[sms>0]))[0][0]+1 # find the minimum index
  n = Nmax-K+1 # conver to n value
  Fp = F - sum(np.arange(K+1,Nmax+1))
  return [int(n),Fp]

def decnr(n,F): # decoder for n'
  nr = n+F-1
  return int(nr)

def decodeIndex(F): # method that converts the index to n and l values
  [l,Fp] = decl(F) # decode the l component
  [n,Fpp] = decn(Fp) # decode the n component
  nr = decnr(n,Fpp) # decode the np component
  return [nr,n,l] # return the (n', n, l) triple for this index
```

```python
# decode each of the important features

important_features = np.array([616, 778])

for ind in important_features:
  [nr,n,l] = decodeIndex(ind)
  print(f"Properties for index {ind}: n'={nr}, n={n}, l={l}")
```
